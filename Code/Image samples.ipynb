```python
img = imread('../../BigData/kaggle-right-whale/right_whale_hunt/imgs/whale_00195/w_2753.jpg')
```

```python
segments_slic = slic(img, n_segments=5, compactness=20, sigma=1)
```

```python
segments_felz = felzenszwalb(img, scale=1000)
```

```python
plt.imshow(mark_boundaries(img, segments_felz))
```

```python
plt.imshow(mark_boundaries(img, segments_slic))
```
